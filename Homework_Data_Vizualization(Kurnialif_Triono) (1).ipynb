Dari data yang diperoleh dari suatu bank karena banyaknya nasabah yang mengakhiri atau menghentikan layanan kartu kreditnya, diperoleh informasi bahwa ada sekitar 16.07% yakni 1627 nasabah (Attrited Customer) yang memilih mengakhiri atau menghentikan layanan kartu kreditnya. Hal ini dapat kita analisa lebih lanjut pada ***Credit Card Customers Dataset by Attrition Flag*** pada bagan dibawah ini:




```python
# group by Attrition_Flag
attrition_group = df.groupby('Attrition_Flag')

# create subplots for each variable
fig, axs = plt.subplots(nrows=3, ncols=3, figsize=(30, 15))
fig.suptitle('Credit Card Customers Dataset by Attrition Flag')

# plot for customer age
sns.histplot(data=df, x='Customer_Age', hue='Attrition_Flag', kde=True, ax=axs[0, 0])
axs[0, 0].set_title('Customer_Age')

# plot for gender
sns.countplot(data=df, x='Gender', hue='Attrition_Flag', ax=axs[0, 1])
axs[0, 1].set_title('Gender')
for p in axs[0, 1].patches:
    axs[0, 1].annotate(f'\n{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=10, color='black', xytext=(0, 13), textcoords='offset points')

# plot for education level
sns.countplot(data=df, x='Education_Level', hue='Attrition_Flag', ax=axs[0, 2])
axs[0, 2].set_title('Education Level')
for p in axs[0, 2].patches:
    axs[0, 2].annotate(f'\n{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=10, color='black', xytext=(0, 13), textcoords='offset points')

# plot for income category
sns.countplot(data=df, x='Income_Category', hue='Attrition_Flag', ax=axs[1, 0])
axs[1, 0].set_title('Income Category')
for p in axs[1, 0].patches:
    axs[1, 0].annotate(f'\n{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=10, color='black', xytext=(0, 13), textcoords='offset points')

# plot for card utilization rate
sns.histplot(data=df, x='Avg_Utilization_Ratio', hue='Attrition_Flag', kde=True, ax=axs[1, 1])
axs[1, 1].set_title('Card Utilization Rate')

# plot for total transaction amount
sns.histplot(data=df, x='Total_Trans_Amt', hue='Attrition_Flag', kde=True, ax=axs[1, 2])
axs[1, 2].set_title('Total Transaction Amount')
for p in axs[1, 2].patches:
    axs[1, 2].annotate(f'\n{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=10, color='black', xytext=(0, 13), textcoords='offset points')

# plot for credit card type
sns.countplot(data=df, x='Card_Category', hue='Attrition_Flag', ax=axs[2, 0])
axs[2, 0].set_title('Card_Category')
for p in axs[2, 0].patches:
    axs[2, 0].annotate(f'\n{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=10, color='black', xytext=(0, 13), textcoords='offset points')

# plot for dependent count
sns.countplot(data=df, x='Dependent_count', hue='Attrition_Flag', ax=axs[2, 1])
axs[2, 1].set_title('Dependent Count ')
for p in axs[2, 1].patches:
    axs[2, 1].annotate(f'\n{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=10, color='black', xytext=(0, 13), textcoords='offset points')

# plot for total transaction CT
sns.histplot(data=df, x='Total_Trans_Ct', hue='Attrition_Flag', kde=True, ax=axs[2, 2])
axs[2, 2].set_title('Total Transaction CT')
for p in axs[2, 2].patches:
    axs[2, 2].annotate(f'\n{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=10, color='black', xytext=(0, 13), textcoords='offset points')

# adjust layout
plt.tight_layout()
```

Dari hasil visualisasi data berdasarkan nasabah yang churned (Attrited Customer), beberapa informasi dan insight yang dapat diperoleh antara lain:
*  Berdasarkan umur nasabah (*Customer Age*): Nasabah yang memilih untuk mengakhiri atau menghentikan layanan kartu kredit (Attrited Customer) cenderung lebih sedikit daripada nasabah yang tetap (Existing Customer). Sebagian besar nasabah yang mengakhiri atau menghentikan berada di usia antara 40-60 tahun.

  Hal ini dapat menjadi pertimbangan bagi perusahaan perbankan dalam mengevaluasi produk dan layanan yang ditawarkan kepada customer direntang usia tersebut. Perusahaan perbankan dapat mempertimbangkan untuk menawarkan produk dan layanan yang lebih sesuai dengan preferensi dan kebutuhan customer, serta meningkatkan strategi retensi customer yang lebih muda agar dapat meminimalkan tingkat churn.
*  Berdasarkan jenis kelamin (*Gender*): Nasabah yang memilih untuk mengakhiri atau menghentikan layanan kartu kredit (Attrited Customer) lebih tinggi pada nasabah wanita(930 nasabah) dibanding nasabah pria.
*  Berdasarkan tingkat pendidikan (*Education Level*): Proporsi customer yang churn (Attrited Customer) cenderung lebih tinggi pada tingkat pendidikan "Graduate"(467 nasabah) daripada yang lainnya.
*  Berdasarkan pendapatan nasabah (*Income Category*): Proporsi customer yang churn (Attrited Customer) cenderung lebih tinggi pada income category `Less than $40K` daripada yang lainnya.
*  Berdasarkan rasio penggunaan rata-rata kartu kredit oleh customer (*Card Utilization Rate*): Jika dilihat tingkat churn paling banyak ada pada rasio penggunaan kartu kredit ada dibawah 0.2, maka dapat diartikan bahwa customer cenderung tidak memanfaatkan kartu kredit mereka secara maksimal, dengan menggunakan kurang dari 20% dari total batas kredit yang tersedia.

  Dampak dari penggunaan kartu kredit yang rendah ini adalah perusahaan perbankan mungkin mengalami penurunan pendapatan dari bunga dan biaya kartu yang biasanya dibebankan pada penggunaan kartu kredit. Namun, di sisi lain, penggunaan kartu kredit yang rendah juga dapat menunjukkan bahwa customer memiliki tingkat pengeluaran yang rendah dan dapat mengelola keuangan mereka dengan baik. Oleh karena itu, perusahaan perbankan dapat mengambil pendekatan yang berbeda dalam menghadapi customer dengan rasio penggunaan kartu yang rendah ini, seperti menawarkan produk kartu kredit dengan fitur dan keuntungan yang lebih menarik untuk meningkatkan penggunaan kartu kredit.
*  Berdasarkan total nilai transaksi finansial yang dilakukan oleh seorang customer pada periode waktu tertentu (*Total_Trans_Amt*): Proporsi customer yang churn (Attrited Customer) cenderung melakukan jumlah nilai transaksi yang lebih rendah(dibawah 2500) daripada yang tidak churn (Existing Customer).
*  Sedangkan berdasarkan jumlah total transaksi yang dilakukan oleh seorang customer pada periode waktu tertentu (*Total_Trans_Ct*): Nasabah dengan jumlah total transaksi dibawah 60 memiliki tingat churn lebih tinggi.
*  Berdasarkan tipe kartu (*Card Category*): Nasabah yang churn (Attrited Customer) cenderung lebih tinggi pada pemegang kartu bertipe "Blue" dan yang kedua "Silver".

  Pada varibel ini dapat membantu perusahaan perbankan dalam membuat strategi penetapan harga dan pemasaran produk kartu kredit yang lebih efektif. Misalnya, perusahaan perbankan dapat menawarkan jenis kartu kredit yang berbeda dengan manfaat dan keuntungan yang berbeda-beda, dan menargetkan segmen customer yang berbeda dengan jenis kartu kredit yang paling sesuai dengan profil dan kebutuhan mereka.
*  Berdasarkan jumlah tanggungan atau anggota keluarga yang bergantung pada seorang nasabah (*Dependent_count*): Nasabah dengan jumlah tanggungan 3 dan 2, memiliki tingkat churn lebih tinggi. Hal ini dapat menunjukkan bahwa nasabah dengan jumlah tanggungan lebih tinggi mungkin memiliki kebutuhan keuangan yang lebih besar dan lebih banyak pengeluaran dalam kehidupan sehari-hari, sehingga mereka membutuhkan kartu kredit untuk membantu mereka memenuhi kebutuhan finansial mereka.

  Sehingga menjadi salah satu faktor dalam mengembangkan strategi pemasaran dan penawaran produk dan layanan perbankan. Misalnya, mereka dapat menawarkan program diskon atau penawaran khusus untuk customer dengan jumlah tanggungan yang lebih tinggi, untuk membantu mereka menghemat uang dan memenuhi kebutuhan finansial mereka yang lebih besar. Namun perlu diperhatikan nasabah dengan jumlah tanggungan yang lebih tinggi mungkin memerlukan batas kredit yang lebih tinggi, namun risiko gagal bayar juga mungkin lebih tinggi karena mereka memiliki lebih banyak tanggungan.

**Kesimpulan:**

Dari data dan informasi tersebut, dapat disimpulkan bahwa target market dari bank untuk mengurangi tingkat churn atau nasabah yang mengakhiri atau menghentikan layanan kartu kredit adalah nasabah perempuan dengan rentang usia 40-60 tahun yang memiliki tingkat pendidikan Graduate dan penghasilan dalam kategori `Less than $40K`. Selain itu, perusahaan juga perlu memperhatikan faktor-faktor lain seperti jenis kartu kredit yang ditawarkan, biaya dan bunga yang kompetitif, dan pelayanan yang baik untuk mempertahankan nasabah agar tidak churn.
